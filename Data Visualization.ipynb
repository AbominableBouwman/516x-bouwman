```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
demographic = pd.read_excel("Demographic Information.xlsx", sheet_name="Form Responses 1")
```

```python
demographic.groupby(['How old are you? ']).size()
sns.histplot(x = 'How old are you? ', data= demographic, hue = 'How old are you? ', legend = False)
```

```python
plt.pie(demographic.groupby(['How old are you? ']).size(), autopct='%.0f%%', labels = set(demographic['How old are you? ']))
plt.title('Age Grouping of Sample Population')
plt.show()
```

```python
plt.pie(demographic.groupby(['What is your biological sex']).size(), autopct='%.0f%%', labels = set(demographic['What is your biological sex']))
plt.title('Sex of Sample Population')
plt.show()
```

```python
df = pd.read_excel('reaction_time_cleaned.xlsx')
df = df[np.abs(df.time - df.time.mean()) <= (3*df.time.std())]
df.shape
```

```python
df.head()
```

```python
ax = sns.boxplot(x='method', y='time', data=df)
ax.set_ylim(0, .75)
plt.title('Reaction Time by Method of Testing')
plt.ylabel('Method of Testing')
plt.xlabel('Reaction Times (Seconds)')
plt.show()
plt.show()
```

```python
ax = sns.boxplot(x='age', y='time', data=df, order= ['14 or younger', '15-24', '25-44','45-64', '65+'], hue = 'method')
ax.set_ylim(0, .65)
plt.title('Reaction Time by Age group sectioned by Testing Method')
plt.xlabel('Method of Testing')
plt.ylabel('Reaction Times (Seconds)')
plt.legend(loc= 'right',  bbox_to_anchor=(1.35, .25), title = 'Method')
plt.show()
```

```python
ax = sns.boxplot(x='method', y='time', data=df, hue = 'age' )
plt.title('Reaction Time by Method of Testing sectioned by Age Groups')
plt.xlabel('Method of Testing')
plt.ylabel('Reaction Times (Seconds)')
plt.legend(loc= 'right',  bbox_to_anchor=(1.35, .25), title = 'Age')
ax.set_ylim(0, .7)
plt.show()
```

```python
ax = sns.boxplot(x='method', y='time', data=df, hue = 'tech_level')
ax.set_ylim(0, .75)
plt.title('Reaction Time by Method of Testing sectioned by Tech Comfort Level')
plt.xlabel('Method of Testing')
plt.ylabel('Reaction Times (Seconds)')
plt.legend(loc= 'right',  bbox_to_anchor=(1.25, .25), title = 'Tech Level')
plt.show()
```
